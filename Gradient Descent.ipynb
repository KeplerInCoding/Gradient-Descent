```python
def gradient_descent(x,y):
    m = b = 0
    i = 1000
    n = len(x)
    a = 0.00001
    
    for j in range(i):
        y0 = m * x + b
        cost = (1/n) * sum((y-y0)**2)
        dm = -(2/n) * sum(x*(y-y0))
        db = -(2/n) * sum(y-y0)
        m = m - a * dm
        b = b - a * db
        print ("m {}, b {}, cost {}".format(m,b,cost))
x= np.array(d.Height)         #though the values of x and y is already assigned
y= np.array(d.Weight)
gradient_descent(x,y)
```
